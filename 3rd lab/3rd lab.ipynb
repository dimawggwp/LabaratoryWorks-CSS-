```python
#1
def even_numbers(n):
    for i in range(1, n + 1):
        if i % 2 == 0:
            if i % 4 == 0:
                yield "кратно 4"
            else:
                yield i


for x in even_numbers(10):
    print(x)
```

```python
#2
def filter_words(words):
    for word in words:
        if len(word) > 4:
            if "а" in word:
                yield "с а"
            else:
                yield word

words = ["кот", "машина", "арбуз", "дом"]

for w in filter_words(words):
    print(w)
```

```python
#3
def infinite_numbers():
    i = 1
    while True:
        if i % 3 == 0 and i % 5 == 0:
            yield "FizzBuzz"
        elif i % 3 == 0:
            yield "Fizz"
        elif i % 5 == 0:
            yield "Buzz"
        else:
            yield i
        i += 1


gen = infinite_numbers()

for eshkere in range(15):
    print(next(gen))
```

```python
#4
def squares(n):
    for i in range(1, n + 1):
        if (i**2) % 2 == 0:
            yield "чётный квадрат"
        else:
            yield i
for x in squares(5):
    print(x)
```

```python
#1
def is_prime(num):
    if num < 2:
        return False
    for j in range(2, int(num**0.5) + 1):
        if num % j == 0:
            return False
    return True

def special_numbers(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            yield "FizzBuzz"
        elif i % 3 == 0:
            yield "Fizz"
        elif i % 5 == 0:
            yield "Buzz"
        elif is_prime(i):
            yield "простое"
        else:
            yield i

for x in special_numbers(15):
    print(x)
```

```python
#2
words = ["кот", "машина", "арбуз", "дом", "ананас"]
res = [(lambda w: (w.upper() + "*" if "а" in w else w.upper()) if len(w) > 4 else "short")(w) for w in words]

print(res)
```

```python
#3
def process_numbers(numbers):
    for num in numbers:
        if num >= 0:
            yield (lambda x: x / 2 if x % 2 == 0 else x * 3 + 1)(num)


numbers = [5, -2, 8, 0, -7, 3]

for x in process_numbers(numbers):
    print(x)
```

```python
#4
students = [("Иван", 85), ("Анна", 72), ("Пётр", 90), ("Мария", 60)]
grade_level = (lambda score: "Отлично" if score >= 90 else "Хорошо" if score >= 70 else "Удовлетворительно")

student_dict = {name: grade_level(score) for name, score in students}

print(student_dict)
```

```python
#5
matrix = [
[1, 2, 3],
[4, 5, 6],
[7, 8, 9]
]
def matrix_transform(matrix):
    for i in matrix:
        for j in i:
            if j%2==0 and j%3==0:
                yield "кратно 6"
            elif j%3==0:
                yield "кратно 3"
            elif j%2==0:
                yield "четное"
            else:
                yield j
for x in matrix_transform(matrix):
    print(x)
```

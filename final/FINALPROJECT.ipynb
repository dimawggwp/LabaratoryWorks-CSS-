```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("music_catalog.xlsx")
print("Первые 5 строк:")
print(df.head())
print("\nРазмер данных (строки, столбцы):")
print(df.shape)
print("\nТипы данных колонок:")
print(df.dtypes)
print("\nКоличество пропусков в каждой колонке:")
print(df.isnull().sum())
```

```python
#2
genre_list = df["Genre"].dropna().unique().tolist()
key_list = df["Key"].dropna().unique().tolist()
print("Жанры:", genre_list)
print("Ключи:", key_list)
```

```python
#3
titles = df["Track_ID"].dropna()
upper_titles = titles.str.upper().tolist()
title_lengths = titles.str.len().tolist()
mean_length = titles.str.len().tolist()

print("Пример верхних названий:", upper_titles[:5])
print("Пример длин:", title_lengths[:5])
print("Средняя длина:", mean_length)
```

```python
#4
top_tracks = df[
    (df["User_Rating"] >= 8) &
    (df["Streams_million"] >= 100)
]
result = top_tracks[["Title", "Artist", "Streams_million", "User_Rating"]].head(10)
print(result)
```

```python
#5
df["total_awards"] = df["Awards_won"] + df["Awards_nominated"]

top_awards = df.sort_values(by="total_awards", ascending=False)

result = top_awards[["Title", "Artist", "Awards_won", "Awards_nominated", "total_awards"]].head(10)

print(result)
```

```python
#6
file_name = "student1_top_tracks.xlsx"
top_tracks.to_excel(file_name, index=False)
print(f"Файл сохранён как {file_name}")

#проверка на корректность
check_df = pd.read_excel(file_name)
print(check_df.head())
```

```python
#7
filtered_tracks = df[df["Title"].str.contains("Song_1")]
print(filtered_tracks[["Title", "Artist", "Genre", "User_Rating"]])
```

```python
#8
top_streams = df.sort_values(by="Streams_million", ascending=False)
print("Топ по стримам:")
print(top_streams[["Title", "Artist", "Streams_million"]].head(10))

short_tracks = df.sort_values(by="Duration_sec", ascending=True)
print("\nСамые короткие треки:")
print(short_tracks[["Title", "Artist", "Duration_sec"]].head(10))
```

```python
#9
df["streams_per_award"] = df["Streams_million"] / (df["total_awards"] + 1)
top5 = df.sort_values(by="streams_per_award", ascending=False)
print(top5[["Title", "Artist", "streams_per_award"]].head(5))
```

```python
#10
def pop_tracks_generator(df):
    for _, row in df.iterrows():
        if row["Genre"] == "Pop":
            yield {
                "Title": row["Title"],
                "Artist": row["Artist"],
                "Genre": row["Genre"],
                "Streams_million": row["Streams_million"]
            }
gen = pop_tracks_generator(df)

for i in range(20):
    try:
        track = next(gen)
        print(track)
    except StopIteration:
        break
```

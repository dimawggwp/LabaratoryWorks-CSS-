```python
file = open("shop_logs.txt", "r")

unique_users = set()
total_buys = 0
total_sum = 0
user_spending = {}

for line in file:
    parts = line.strip().split(";")
    user_id = parts[1]
    action = parts[2]

    unique_users.add(user_id)

    if action == "BUY":
        amount = int(parts[3])
        total_buys += 1
        total_sum += amount

        if user_id not in user_spending:
            user_spending[user_id] = amount
        else:
            user_spending[user_id] += amount

file.close()

max_user = ""
max_spent = 0
for user in user_spending:
    if user_spending[user] > max_spent:
        max_spent = user_spending[user]
        max_user = user

average_check = total_sum / total_buys if total_buys > 0 else 0

# Выводим результаты на экран
print("Уникальных пользователей:", len(unique_users))
print("Всего покупок:", total_buys)
print("Общая сумма:", total_sum)
print("Самый активный покупатель:", max_user)
print("Средний чек:", average_check)
```

```python
#2
import csv
file = open("employees.csv", "r")
dep = {}
emp = []
total_salary = 0
count = 0
header = file.readline()
for line in file:
    parts = line.strip().split(",")

    user_name = parts[0]
    user_part = parts[1]
    salary = int(parts[2])

    emp.append(parts)

    total_salary += salary
    count += 1

    if user_part not in dep: #если отдела нет я добавил пустой списко что бы туда добавить инфу
        dep[user_part] = []
    dep[user_part].append(salary)

file.close()

#ср.зп
if count > 0:
    avg_salary = total_salary / count
else:
    avg_salary = 0


#ср.зп по отделц
avg_salary_part = {}

for part in dep:
    sum_dep = 0
    for s in dep[part]:
        sum_dep += s
    avg_salary_part[part] = sum_dep / len(dep[part])

#ОТдел с ссамым высоким срзп
highest_part = ""
highest_avg = 0
for part in avg_salary_part:
    if avg_salary_part[part] > highest_avg:
        highest_avg = avg_salary_part[part]
        highest_part = part

#самый много получающий лев
max_salary = 0
max_employee = ""
for e in emp:
    salary = int(e[2]) #e[2] ну третий индекс тоесть зп
    if salary > max_salary:
        max_salary = salary
        max_employee = e[0]

#работники высше ср зп (чемв кз :( )
high_salary = []

for e in emp:
    if int(e[2]) > avg_salary:
        high_salary.append(e[0])

#фуууух все условия выполнены
print("Ср зп:", avg_salary)
print("Ср по отделам:", avg_salary_part)
print("Отдел с макс зп:", highest_part)
print("Хищник с макс зп:", max_employee)
print("Хищники высше ср зп:", high_salary)
```

```python
#3
import json

with open("orders.json", "r", encoding="utf-8") as file:
    orders = json.load(file)

total_revenue = 0
user_orders = {}
total_items = 0
max_order_total = 0
top_user = ""
item_count = {}

for order in orders:
    user = order["user"]
    items = order["items"]
    total = order["total"]

    #Общ сумма ордеров
    total_revenue += total

    #кто скок купил
    if user not in user_orders:
        user_orders[user] = 1
    else:
        user_orders[user] += 1

    # общ колюво товаров
    total_items += len(items)

    #Тигр с самым дорогим ордером
    if total > max_order_total:
        max_order_total = total
        top_user = user

    # КолюВо каждого товара
    for item in items:
        if item not in item_count:
            item_count[item] = 1
        else:
            item_count[item] += 1

# свамый популяр товар
most_popular_item = ""
max_item_count = 0
for item in item_count:
    if item_count[item] > max_item_count:
        max_item_count = item_count[item]
        most_popular_item = item

# Создадим словарь для суммари
summary = {
    "total_revenue": total_revenue,
    "top_user": top_user,
    "most_popular_item": most_popular_item,
    "total_orders": len(orders)
}

# Выводим результаты
print("Общ колво ордеров:", total_revenue)
print("Колво заордеров по юзерам:", user_orders)
print("Общ колво товаров:", total_items)
print("Лев с самым дорогим заказом:", top_user)
print("Самый популярный товар:", most_popular_item)

#Сохраненине суммари в джейсон
with open("summary.json", "w", encoding="utf-8") as f:
    json.dump(summary, f, indent=2)
```

```python
#4
import csv
import json

file = open("transactions.csv", "r")
reader = csv.reader(file)

header = next(reader)
suspicious_transactions = []
suspicious_users = set()
user_operations = {}
total_suspicious_sum = 0

for row in reader:
    user_id = row[0]
    amount = int(row[1])

    if user_id not in user_operations:
        user_operations[user_id] = 1
    else:
        user_operations[user_id] += 1

    if amount > 500000:
        suspicious_transactions.append((user_id, amount))
        suspicious_users.add(user_id)
        total_suspicious_sum += amount

file.close()

for user in user_operations:
    if user_operations[user] > 3:
        suspicious_users.add(user)

report = open("fraud_report.txt", "w")

report.write("Подозрительных транзакций: " + str(len(suspicious_transactions)) + "\n")
report.write("Подозрительных пользователей: " + str(len(suspicious_users)) + "\n")
report.write("Список пользователей: " + ", ".join(suspicious_users) + "\n")
report.write("Общая сумма подозрительных операций: " + str(total_suspicious_sum) + "\n")

report.close()
json_file = open("fraud_users.json", "w")
json.dump(list(suspicious_users), json_file, indent=4)
json_file.close()
print("4 это гг конечно, волосы даже поседели")
```
